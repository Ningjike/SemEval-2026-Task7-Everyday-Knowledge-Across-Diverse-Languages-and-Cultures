```python
prompt_format = '''### Instruction: Answer the following question in {target_language}, as a local person from {country}. Be concise and culturally accurate.
### Question: {question}
### Response:
<think>
'''

def build_prompt(lang_region: str, question: str, locale_map=LANG_RE_INFO) -> str:
    info = locale_map.get(lang_region)
    if info is None:
        target_language = "the requested language"
        country = "the relevant region"
    else:
        target_language = info["lang_name"]
        country = info["region_name"]
    
    return prompt_format.format(
        target_language=target_language,
        country=country,
        question=question
    )
```

```python
prompt1 = build_prompt("ms-SG", "Apakah akronim lazim untuk flat perumahan awam di Singapura?")
print(prompt1)
response = pipeline(prompt1)
print(response[0]["generated_text"])
```

```python
from tqdm.auto import tqdm
import re
from datasets import Dataset

def pipeline_with_progress(pipeline, prompts, batch_size=4, **kwargs):
    total_batches = (len(prompts) + batch_size - 1) // batch_size
    
    # 分批处理并显示进度
    all_responses = []
    for batch_idx in tqdm(range(total_batches), desc="inference"):
        start_idx = batch_idx * batch_size
        end_idx = min((batch_idx + 1) * batch_size, len(prompts))
        batch_prompts = prompts[start_idx:end_idx]
        
        batch_responses = pipeline(
            batch_prompts,
            batch_size=len(batch_prompts),
            **kwargs
        )
        all_responses.extend(batch_responses)
    
    return all_responses
    
prompt_list = []
id_list = []

for _, row in test.iterrows():
    prompt = build_prompt(
        lang_region=row["lang_region"],
        question=row["question"]
    )
    prompt_list.append(prompt)
    id_list.append(row["id"])

responses = pipeline_with_progress(
    pipeline,
    prompt_list,
    batch_size=4,
    do_sample=False,
    pad_token_id=pipeline.tokenizer.eos_token_id,
    max_new_tokens=300,
)
test_pred = []
for i, resp in enumerate(responses):
    generated = resp[0]["generated_text"]
    prompt = prompt_list[i]
    if generated.startswith(prompt):
        text_after_prompt  = generated[len(prompt):].strip()
    else:
        text_after_prompt  = generated.strip()

    match = re.search(r"</think>\s*(.*)", text_after_prompt , re.DOTALL)
    if match:
        ans = match.group(1).strip()
    else:
        ans = text_after_prompt 
        
    ans = ans.split("\n")[0].strip()
    test_pred.append(ans)

```

```python
result_output = pd.DataFrame({"id": id_list, "ans": test_pred})
result_output.to_csv("/kaggle/working/instruction_learning.csv", index=False)
```
